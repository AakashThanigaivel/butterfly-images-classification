Build a Convolution Neural Network to classify 6 different species of Butterfly images.

<u>**Dataset_Folder Name**</u>:  Butterflies_classification_dataset

<u> **Conditions to consider** </u>

- Parameters should not cross 300000

- Should not use more than 4 layers (except input and output, including convolution and dense layers)

- Use Adam Optimizer

```python
# Importing Required Libraries
import tensorflow as tf
from keras.preprocessing.image import ImageDataGenerator
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Conv2D, MaxPool2D, Flatten, Dense, GlobalAveragePooling2D, Dropout, MaxPooling2D
from tensorflow.keras.optimizers import Adam
from tensorflow.keras.preprocessing.image import ImageDataGenerator
from tensorflow.keras.callbacks import EarlyStopping, ModelCheckpoint
from sklearn.metrics import classification_report,confusion_matrix
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import os
os.environ["CUDA_DEVICE_ORDER"]="PCI_BUS_ID"
os.environ["CUDA_VISIBLE_DEVICES"]="0"
```

```python
# Train and Test File Directory
train_dir=r"C:\Users\saisc\Downloads\DL-CT-3\train\train"
test_dir=r"C:\Users\saisc\Downloads\DL-CT-3\test\test"
```

```python
os.listdir()
```

```python
# Initialize Generator
gen = ImageDataGenerator(
    rescale=1./255,
    horizontal_flip=True, 
    rotation_range=20,
    validation_split=0.2
)

# Load Data
train_ds = gen.flow_from_directory(
    train_dir,
    class_mode='categorical',
    shuffle=True,
    batch_size=32,
    target_size=(128,128),
    subset="training"
)

valid_ds = gen.flow_from_directory(
    test_dir,
    class_mode='categorical',
    shuffle=True,
    batch_size=32,
    target_size=(128,128),
    subset="validation")
```

```python
# Different Classes of Dependent Variable
class_name=sorted(os.listdir(train_dir))
class_name
```

```python
# Defining the Graph Function
def show_image(image, title=None):
    plt.imshow(image)
    plt.title(title)
    plt.axis('off')
```

```python
# Building the Model
model = Sequential()
model.add(Conv2D(8,kernel_size=(9,9),padding = 'same',activation ='relu', input_shape = (128,128,3)))
model.add(MaxPooling2D(pool_size=(7,7)))
model.add(Flatten())
model.add(Dense(70,activation='relu'))
model.add(Dense(20,activation='relu'))
model.add(Dense(6, activation = "softmax"))
```

```python
model.compile(loss='categorical_crossentropy',optimizer='adam',metrics=['accuracy'])
```

```python
history = model.fit(
  train_ds,
  validation_data=valid_ds,
  epochs=10,
  steps_per_epoch=len(train_ds),
  validation_steps=len(valid_ds)
)
```

```python
model.summary()
```

```python
#import matplotlib.pyplot as plt
plt.plot(history.history["loss"])
plt.plot(history.history['accuracy'])
plt.plot(history.history['val_loss'])
plt.plot(history.history['val_accuracy'])
plt.title("Model accuracy")
plt.ylabel("Accuracy")
plt.xlabel("Epoch")
plt.legend(["Loss","Accuracy","Validation Loss","Validation Accuracy"])
plt.show()
```

```python
# Building the Model
model = Sequential()
model.add(Conv2D(10,kernel_size=(9,9),padding = 'same',activation ='relu', input_shape = (128,128,3)))
model.add(MaxPooling2D(pool_size=(5,5)))
model.add(Flatten())
model.add(Dense(70,activation='relu'))
model.add(Dense(20,activation='relu'))
model.add(Dense(6, activation = "softmax"))
```

```python
model.compile(loss='categorical_crossentropy',optimizer='adam',metrics=['accuracy'])
```

```python
checkpoint = ModelCheckpoint(r"C:\Users\saisc\Downloads\DL-CT-3\save_best.h5", monitor='val_loss', verbose = 1, save_best_only=True)
```

```python
history = model.fit(
  train_ds,
  validation_data=valid_ds,
  epochs=20,
  steps_per_epoch=len(train_ds),
  validation_steps=len(valid_ds),callbacks=checkpoint
)
```
